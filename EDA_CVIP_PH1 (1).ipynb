importing required libraries


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

read the dataset

```python
tr=pd.read_csv("Terrorism.csv", encoding='ISO-8859-1')
```

customizing the columns

renaming required columns and using only them

```python
tr=tr.rename(columns={'iyear':'year', 'imonth':'month', 'iday':'day', 'country_txt':'Country','region_txt':'Region','provstate':'State', 'attacktype1_txt':'attack', 'targtype1_txt':'targettype','targsubtype1_txt':'target', 'gname':'gang', 'weaptype1_txt':'weapon'})
```

```python
tr=tr[['year','month','day','Country','Region','State','city','attack','targettype','target','gang','weapon']]
```

fixing the missing values

```python
tr['State']=tr['State'].fillna('unknown')
tr['target']=tr['target'].fillna('unknown')
tr
```

```python
tr.describe()
```

Visualization

```python
plt.figure(figsize=(15,6))
plt.xticks(rotation=30)
plt.title('Based on Attack type')
sns.countplot(x=tr['attack'])
```

```python
plt.figure(figsize=(15,6))
plt.xticks(rotation=90)
plt.title('Attacks by month')
sns.countplot(x=tr['month'])
```

```python
plt.figure(figsize=(15,6))
plt.xticks(rotation=90)
plt.title('Attacks on different days')
sns.countplot(x=tr['day'])
```

```python
plt.figure(figsize=(15,6))

plt.title('Types of weapon used')
sns.countplot(x=tr['weapon'])
```

```python
pd.crosstab(tr.year, tr.Region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()

```

```python
plt.figure(figsize=(15,6))
plt.xticks(rotation=90)
plt.title('Types of Target People')
sns.countplot(x=tr['targettype'])
```
